```python
import pandas as pd
import numpy as np
import datetime, time
import json
```

# Initialisation

```python
def window(year1, month1, day1, year2, month2, day2, df_):
    date1 = datetime.datetime(year=year1, month=month1, day=day1).timestamp()
    date2 = datetime.datetime(year=year2, month=month2, day=day2).timestamp()
    wd = df_[(date1 <= df_csv.Timestamp) & (df_.Timestamp <= date2)]
    return wd
```

```python
df = pd.DataFrame(
    columns=[
        "Name", "Symbol", "Date", "Watching",
        "Data2018", "Rate", 
        "Price", "Volume", "FullyDilluted", "CirculatingSupply"],
)
```

# Main

```python
not_available = []
for i in range(1, 11131):
    try:
        # json data
        js = json.load(open(f"detail/{i}.json"))
        data = js["data"]
        id = data["id"]
        name = data["name"]
        symbol = data["symbol"]
        date = data["dateAdded"]
        watch = data["watchCount"]
        # csv data
        df_csv = pd.read_csv(f"csv/{i}.csv")
        wd2018 = window(2018, 3, 1, 2018, 11, 1, df_csv)
        wd2021 = window(2021, 1, 1, 2021, 7, 1, df_csv)
        price2021 = wd2021.Price.mean()
        price2018 = wd2018.Price.mean()
        if len(wd2018) > 0 and price2018 != 0:
            data2018 = True
            rate = price2021 / price2018
            wdAnalyse2018 = window(2017, 11, 1, 2018, 3, 1, df_csv)
            price = wdAnalyse2018.Price.mean()
            volume = wdAnalyse2018.Volume.mean()
            fullyDilluted = wdAnalyse2018.FullyDilluted.mean()
            circulatingSupply = wdAnalyse2018.CirculatingSupply.mean()
        else:
            data2018 = False
            rate, price, volume, fullyDilluted, circulatingSupply = [None]*5
        df.loc[id] = [
            name, symbol, date, watch,
            data2018, rate, 
            price, volume, fullyDilluted, circulatingSupply
        ]
    except:
        not_available.append(i)
print(not_available)
```

```python
df.to_csv("summary.csv")
```
